```python
import numpy as np
```

```python
def load_file_data(path):
    file = open(path, 'r')
    data = []
    for line in file:
        data.append(line.strip("\n").split())
    return data
```

```python
mapping = dict(load_file_data('data/fb15k/id2entity.sme.txt'))
```

```python
elements = np.empty(14952, dtype='object')
for k in mapping:
    if (int(k) > 14951):
        continue
    elements[int(k)] = mapping[k]
```

```python
file = open('data/fb15k/predictions.transe.txt', 'r')
out = open('data/fb15k/predictions.sme.txt', 'w')
n = 300

for line in file:
    data = line.strip("\n").split(" ")
    scores = zip(elements[np.argsort(data[3:])][n::-1], np.sort(data[3:])[n::-1])
    joined = []
    for a, b in scores:
        if a == elements[14951]:
            continue
        joined.append(a + " " + b)
    
    out.write(" ".join(joined))
    out.write("\n")
        
file.close()
out.close()
```
